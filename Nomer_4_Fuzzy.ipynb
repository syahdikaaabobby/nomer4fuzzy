```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Definisikan rentang nilai untuk variabel input dan output
x_sensor1 = np.arange(0, 1.1, 0.1)
x_sensor2 = np.arange(0, 1.1, 0.1)
x_output = np.arange(0, 1.1, 0.1)

# Tentukan fungsi keanggotaan untuk masing-masing variabel
sensor1_low = fuzz.trimf(x_sensor1, [0, 0.5, 1])
sensor1_medium = fuzz.trimf(x_sensor1, [0, 0.5, 1])
sensor1_high = fuzz.trimf(x_sensor1, [0, 0.5, 1])

sensor2_weak = fuzz.trimf(x_sensor2, [0, 0.5, 1])
sensor2_medium = fuzz.trimf(x_sensor2, [0, 0.5, 1])
sensor2_strong = fuzz.trimf(x_sensor2, [0, 0.5, 1])

output_very_low = fuzz.trimf(x_output, [0, 0.25, 0.5])
output_low = fuzz.trimf(x_output, [0, 0.25, 0.5])
output_high = fuzz.trimf(x_output, [0.25, 0.5, 0.75])
output_very_high = fuzz.trimf(x_output, [0.5, 0.75, 1])

# Plotting fungsi keanggotaan
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(x_sensor1, sensor1_low, 'b', label='Low')
plt.plot(x_sensor1, sensor1_medium, 'g', label='Medium')
plt.plot(x_sensor1, sensor1_high, 'r', label='High')
plt.title('Membership Function for Sensor 1')
plt.xlabel('Sensor 1 Value')
plt.ylabel('Membership Degree')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(x_sensor2, sensor2_weak, 'b', label='Weak')
plt.plot(x_sensor2, sensor2_medium, 'g', label='Medium')
plt.plot(x_sensor2, sensor2_strong, 'r', label='Strong')
plt.title('Membership Function for Sensor 2')
plt.xlabel('Sensor 2 Value')
plt.ylabel('Membership Degree')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(x_output, output_very_low, 'b', label='Very Low')
plt.plot(x_output, output_low, 'g', label='Low')
plt.plot(x_output, output_high, 'r', label='High')
plt.plot(x_output, output_very_high, 'y', label='Very High')
plt.title('Membership Function for Output')
plt.xlabel('Output Value')
plt.ylabel('Membership Degree')
plt.legend()

plt.tight_layout()
plt.show()
```
